```python
# to define a function
def duration_split(col):
    hours =0
    minutes = 0
    col=col.split()
    
    if len(col)==1:
        col = ''.join(col)
        if col.endswith('h'):
                hours = col[:-1]
                
        else:
            minutes = col[:-1]
                
    else:
        hours = col[0].strip('h')
                
        minutes =col[1].strip('m')
    return hours, minutes
```

```python
df['Duration'].apply(lambda x: duration_split(x))
```

# to convert it to mins, change the datatype to int and multiply by 60
NB: use \ to indicate continuations

```python
df['Duration'].apply(lambda x: duration_split(x)[0]).astype(int)*60 + \
df['Duration'].apply(lambda x: duration_split(x)[1]).astype(int)
```

```python
# then save it back to DataFrame
df['Duration']=df['Duration'].apply(lambda x: duration_split(x)[0]).astype(int)*60 + \
df['Duration'].apply(lambda x: duration_split(x)[1]).astype(int)
```
